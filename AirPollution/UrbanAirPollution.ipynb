# Modeling

## Scaling 

```python
from sklearn.preprocessing import StandardScaler 

X_scaled = StandardScaler().fit_transform(x)
X = pd.DataFrame(X_scaled, columns=x.columns)
```

```python
X.head(3)
```

## Cross Validation with Different Models

```python
from math import sqrt
from sklearn import metrics
from sklearn.metrics import SCORERS
```

```python
from sklearn.model_selection import cross_validate 

def evaluate_model(estimator):
    cv_results = cross_validate(estimator, X=X, y=target_var,
                                scoring= "neg_mean_squared_error", n_jobs=-1, cv=5, return_train_score=True)
    return np.round(np.sqrt(pd.DataFrame(cv_results).abs().mean()),2).to_dict()
##going to be using MSE bc its a continuous variable problem!
```

```python
RESULTS = {}
```

### Basic Linear Models

```python
from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
```

```python
lr  = LinearRegression()
dtree   = DecisionTreeRegressor()
```

```python
RESULTS["linear reg"] = evaluate_model(lr)
RESULTS["tree"] = evaluate_model(dtree)

pd.DataFrame.from_dict(RESULTS).T
```

**We can see from the MSE scores that the error for the Linear Regression model is rather big.**

### Ensemble Models

```python
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor 
```

```python
bagging_100 = BaggingRegressor(n_estimators=100)
RESULTS["bagging_100"] = evaluate_model(bagging_100)
```

```python
randomforest = RandomForestRegressor(n_estimators=100)
RESULTS["randomforest"] = evaluate_model(randomforest)
```

```python
pd.DataFrame.from_dict(RESULTS).T
```

**We can see that the MSE decreased so these models are better than the basic linear models**

```python
import xgboost
```

```python
from xgboost import XGBRegressor
```

```python
xg_500 = XGBRegressor(n_estimators=500)
RESULTS['xgboost_500'] = evaluate_model(xg_500)
```

```python
from lightgbm import LGBMRegressor
```

```python
lgb = LGBMRegressor(n_estimators=500, max_depth=-1)
RESULTS["lightgbm_500"] = evaluate_model(lgb)
```

```python
pd.DataFrame.from_dict(RESULTS).T
```

**The models which seem to yield the best mse scores are the xgboost and lightgbm models therefore I will use them for my modeling below!**
